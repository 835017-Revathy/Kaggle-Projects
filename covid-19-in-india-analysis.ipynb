```python
covid_df.drop(["Sno","Time","ConfirmedIndianNational","ConfirmedForeignNational"], inplace=True,axis=1)
```

```python
covid_df.head()
```

```python
covid_df['Date']= pd.to_datetime(covid_df['Date'],format='%Y-%m-%d')
```

```python
covid_df.head()
```

```python
#Active Cases
covid_df['Active_Cases'] = covid_df['Confirmed'] - (covid_df['Cured']+covid_df['Deaths'])
covid_df.tail()
```

```python
statewise = pd.pivot_table(covid_df,values=["Confirmed","Deaths","Cured"],
                           index = "State/UnionTerritory",aggfunc = max)
```

```python

#Finding Recovery Rate

statewise["Recovery"] = statewise["Cured"]*100/statewise["Confirmed"]
```

```python
#Finding Mortality Rate
statewise["Mortality"] = statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise = statewise.sort_values(by="Confirmed",ascending=False)
```

```python
statewise.style.background_gradient(cmap="cubehelix")
```

```python
covid_df.rename(columns ={"State/UnionTerritory":"State"},inplace =True)
covid_df.head()
```

```python
covid_df.head()
```

```python
top_10_active  = covid_df.groupby("State").max()['Active_Cases'].to_frame("Active_Cases")
top_10_active = top_10_active.sort_values("Active_Cases",ascending=False)[:10]
top_10_active
```

```python
##Top Ten states based on active cases
fig = plt.figure(figsize=(12,5))
plt.title("Top 10 States with Active Cases India")
x = sns.barplot(data = top_10_active,y=top_10_active.Active_Cases,x=top_10_active.index,linewidth =2,edgecolor='Black')
plt.xlabel("States")
plt.ylabel("Active Cases")
plt.show()
```

```python
top_10_deaths  = covid_df.groupby("State").max()['Deaths'].to_frame("Deaths")
top_10_deaths = top_10_deaths.sort_values("Deaths",ascending=False)[:10]
top_10_deaths
```

```python
##Top Ten states based on Deaths Cases
fig = plt.figure(figsize=(14,5))
plt.title("Top 10 States with Deaths Cases India")
x = sns.barplot(data = top_10_deaths,y=top_10_deaths.Deaths,x=top_10_deaths.index,linewidth =2,edgecolor='Black')
plt.xlabel("States")
plt.ylabel("Deaths")
plt.show()
```

```python
#Remove rows where state is mentioned as India
vaccine = vaccine_df[vaccine_df.State!='India']
vaccine
```

```python
vaccine.rename(columns ={"Total Individuals Vaccinated":"Total"},inplace =True)
vaccine.head()
```

```python
#To Find The states with Most Vaccinated 
vacc_max = vaccine.groupby("State")["Total"].sum().to_frame("Total")
vacc_max = vacc_max.sort_values("Total",ascending=False)[:5]
vacc_max
```

```python
fig = plt.figure(figsize=(6.5,5))
plt.title("Top 5 Most Vaccinated States in India")
x = sns.barplot(data = vacc_max.iloc[:10],y=vacc_max.Total,x=vacc_max.index,linewidth =2,edgecolor='Black')
plt.xlabel("States")
plt.ylabel("Vaccination Count")
plt.show()
```

```python
#To Find The states with Least Vaccinated 
vacc_least = vaccine.groupby("State")["Total"].sum().to_frame("Total")
vacc_least = vacc_least.sort_values("Total",ascending=True)[:5]
vacc_least
```

```python
fig = plt.figure(figsize=(13,5))
plt.title("Top 5 Most Vaccinated States in India")
x = sns.barplot(data = vacc_least.iloc[:10],y=vacc_least.Total,x=vacc_least.index,linewidth =2,edgecolor='Black')
plt.xlabel("States")
plt.ylabel("Vaccination Count")
plt.show()
```
