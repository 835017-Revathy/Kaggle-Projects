```python
covid_df.drop(["Sno","Time","ConfirmedIndianNational","ConfirmedForeignNational"], inplace=True,axis=1)
```

```python
covid_df.head()
```

```python
covid_df['Date']= pd.to_datetime(covid_df['Date'],format='%Y-%m-%d')
```

```python
covid_df.head()
```

```python
#Active Cases
covid_df['Active_Cases'] = covid_df['Confirmed'] - (covid_df['Cured']+covid_df['Deaths'])
covid_df.tail()
```

```python
statewise = pd.pivot_table(covid_df,values=["Confirmed","Deaths","Cured"],
                           index = "State/UnionTerritory",aggfunc = max)
```

```python

#Finding Recovery Rate

statewise["Recovery"] = statewise["Cured"]*100/statewise["Confirmed"]
```

```python
#Finding Mortality Rate
statewise["Mortality"] = statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise = statewise.sort_values(by="Confirmed",ascending=False)
```

```python
statewise.style.background_gradient(cmap="cubehelix")
```

```python
#Remove rows where state is mentioned as India
vaccine = vaccine_df[vaccine_df.State!='India']
vaccine
```

```python
#Most  Vaccinated State
max_vac = vaccine.groupby(by='State')['Total'].sum().to_frame('Total')
max_vac = max_vac.sort_values('Total',ascending=False)[:5]
max_vac
```

```python
fig = plt.figure(figsize=(10,5))
plt.title("Most Vaccinated cities",size=10)
x = sns.barplot(x=max_vac.index,y=max_vac.Total,data=max_vac.iloc[:10],linewidth=2,edgecolor='black')
plt.xlabel("State")
plt.ylabel("Vaccination")
plt.show()
```

```python
least_vac = vaccine.groupby(by='State')['Total'].sum().to_frame('Total')
least_vac = least_vac.sort_values('Total',ascending=True)[:5]
least_vac
```

```python
fig = plt.figure(figsize=(12,6))
plt.title("Least Vaccinated cities",size=10)
x = sns.barplot(x=least_vac.index,y=least_vac.Total,data=least_vac,linewidth=2,edgecolor='black')
plt.xlabel("State")
plt.ylabel("Vaccination")
plt.show()
```
